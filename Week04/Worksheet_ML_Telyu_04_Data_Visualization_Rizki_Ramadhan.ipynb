Beri sample kode python untuk dapat menampilkan persentase dari data bertipe kategorikal dalam bentuk pie chart, sehingga saya bisa copy paste ke google colab.

````python

# Menampilkan persentase data dari kolom bertipe kategorikal
categorical_columns = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']

# Membuat pie chart untuk setiap kolom kategorikal
for column in categorical_columns:
    plt.figure(figsize=(6,6))  # Ukuran pie chart
    df[column].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
    plt.title(f'Pie chart of {column}')
    plt.ylabel('')  # Menghilangkan label pada sumbu y
    plt.show()
````

Prompt:

Beri kode python untuk melihat jumlah siswa yang fail dan pass untuk masing-masing pelajaran (math, reading, writing score). Dengan minimal nilai yang dianggap pass adalah 75!

````python


# Menambahkan kolom baru untuk menentukan apakah siswa pass atau fail berdasarkan nilai 75
df['math_result'] = df['math score'].apply(lambda x: 'Pass' if x >= 75 else 'Fail')
df['reading_result'] = df['reading score'].apply(lambda x: 'Pass' if x >= 75 else 'Fail')
df['writing_result'] = df['writing score'].apply(lambda x: 'Pass' if x >= 75 else 'Fail')

# Menggabungkan hasil pass/fail dari ketiga subjek menjadi satu DataFrame baru untuk visualisasi
result_df = pd.DataFrame({
    'Subject': ['Math']*len(df) + ['Reading']*len(df) + ['Writing']*len(df),
    'Result': list(df['math_result']) + list(df['reading_result']) + list(df['writing_result'])
})

# Membuat countplot untuk melihat jumlah siswa yang pass dan fail di tiap pelajaran
plt.figure(figsize=(10,6))
sns.countplot(x='Subject', hue='Result', data=result_df, palette='Set2')
plt.title('Pass vs Fail for Math, Reading, and Writing')
plt.ylabel('Number of Students')
plt.xlabel('Subject')

# Menampilkan grafik
plt.show()
````

Insight:

siswa terbanyak yang mengalami fail dalam mate pelajaran dengan KKM 75 adalah pelajaran Math
Jumlah siswa yang Fail dalam Math : 700
Jumlah siswa yang Fail dalam Reading : 650
Jumlah siswa yang fail dalam Writing : 675

Jumlah siswa yang Pass dalam Math : 300
Jumlah siswa yang Pass dalam Reading : 390
Jumlah siswa yang Pass dalam Writing : 360

Prompt:

Beri kode python yang menunjukkan visualisasi nilai math, reading, dan writing siswa, dengan melihat pengaruh dari gender, race/ethnicity, parental level of education, lunch, test preparation course-nya!

````python
# Mengatur ukuran keseluruhan plot
plt.figure(figsize=(18, 12))

# Plot 1: Pengaruh Gender terhadap nilai Math
plt.subplot(2, 3, 1)
sns.boxplot(x='gender', y='math score', data=df, palette='Pastel1')
plt.title('Pengaruh Gender terhadap Math Score')

# Plot 2: Pengaruh Race/Ethnicity terhadap nilai Math
plt.subplot(2, 3, 2)
sns.boxplot(x='race/ethnicity', y='math score', data=df, palette='Pastel2')
plt.title('Pengaruh Race/Ethnicity terhadap Math Score')

# Plot 3: Pengaruh Parental Level of Education terhadap nilai Math
plt.subplot(2, 3, 3)
sns.boxplot(x='parental level of education', y='math score', data=df, palette='Set2')
plt.xticks(rotation=45)
plt.title('Pengaruh Parental Education terhadap Math Score')

# Plot 4: Pengaruh Lunch terhadap nilai Reading
plt.subplot(2, 3, 4)
sns.boxplot(x='lunch', y='reading score', data=df, palette='Pastel1')
plt.title('Pengaruh Lunch terhadap Reading Score')

# Plot 5: Pengaruh Test Preparation Course terhadap nilai Reading
plt.subplot(2, 3, 5)
sns.boxplot(x='test preparation course', y='reading score', data=df, palette='Set2')
plt.title('Pengaruh Test Preparation Course terhadap Reading Score')

# Plot 6: Pengaruh Race/Ethnicity terhadap Writing Score
plt.subplot(2, 3, 6)
sns.boxplot(x='race/ethnicity', y='writing score', data=df, palette='Pastel2')
plt.title('Pengaruh Race/Ethnicity terhadap Writing Score')

# Menampilkan plot
plt.tight_layout()
plt.show()
````

Insight :

Pengaruh Gender terhadap math score:

Laki-laki cenderung memiliki nilai math score sedikit lebih tinggi dibandingkan dengan perempuan.
Median  untuk laki-laki berada lebih tinggi dibandingkan dengan perempuan, menunjukkan performa matematik yang lebih baik pada umumnya.
Rentang nilai untuk laki-laki juga sedikit lebih lebar, dengan siswa laki-laki memiliki nilai yang lebih bervariasi dari nilai rendah hingga tinggi.
Terdapat beberapa outliers atau nilai yang sangat rendah baik pada kelompok laki-laki maupun perempuan di bawah nilai 20.

Pengaruh Race terhadap math score:

Group E tampaknya memiliki median math score tertinggi dibandingkan kelompok lainnya, diikuti oleh Group D.
Group A memiliki median math score yang lebih rendah dibandingkan kelompok lainnya, menunjukkan bahwa siswa dari grup ini cenderung memiliki performa yang lebih rendah dalam matematika.

Pengaruh Parental Education terhadap math score:

Siswa dengan orang tua berpendidikan "master's degree" memiliki median math score tertinggi dibandingkan dengan kategori lainnya. Hal ini menunjukkan bahwa siswa yang orang tuanya memiliki pendidikan lebih tinggi cenderung memiliki performa yang lebih baik dalam matematika.
Siswa dengan orang tua berpendidikan "some high school" memiliki median math score terendah, menunjukkan performa yang relatif lebih rendah dalam matematika.

Prompt: Beri kode untuk melihat nilai rata-rata seluruh mata pelajaran (math, reading writing) dari setiap siswa dan lihat korelasinya dengan masing-masing kategori

````python

# Menambahkan kolom baru untuk nilai rata-rata
df['average score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)

# Mengatur ukuran keseluruhan plot
plt.figure(figsize=(18, 12))

# Plot 1: Korelasi antara gender dengan rata-rata nilai
plt.subplot(2, 3, 1)
sns.boxplot(x='gender', y='average score', data=df, palette='Pastel1')
plt.title('Korelasi Gender dengan Rata-rata Nilai')

# Plot 2: Korelasi antara race/ethnicity dengan rata-rata nilai
plt.subplot(2, 3, 2)
sns.boxplot(x='race/ethnicity', y='average score', data=df, palette='Pastel2')
plt.title('Korelasi Race/Ethnicity dengan Rata-rata Nilai')

# Plot 3: Korelasi antara parental level of education dengan rata-rata nilai
plt.subplot(2, 3, 3)
sns.boxplot(x='parental level of education', y='average score', data=df, palette='Set2')
plt.xticks(rotation=45)
plt.title('Korelasi Parental Education dengan Rata-rata Nilai')

# Plot 4: Korelasi antara lunch dengan rata-rata nilai
plt.subplot(2, 3, 4)
sns.boxplot(x='lunch', y='average score', data=df, palette='Pastel1')
plt.title('Korelasi Lunch dengan Rata-rata Nilai')

# Plot 5: Korelasi antara test preparation course dengan rata-rata nilai
plt.subplot(2, 3, 5)
sns.boxplot(x='test preparation course', y='average score', data=df, palette='Set2')
plt.title('Korelasi Test Preparation Course dengan Rata-rata Nilai')

# Menampilkan semua plot dengan layout yang rapi
plt.tight_layout()
plt.show()
````

Insight:

Korelasi gender dengan rata-rata math score:

Perempuan tampaknya memiliki median rata-rata nilai yang sedikit lebih tinggi dibandingkan dengan laki-laki. Hal ini menunjukkan bahwa secara umum, siswa perempuan cenderung mendapatkan nilai rata-rata yang lebih baik.

Korelasi Race dengan rata-rata match score:

Group E memiliki median rata-rata nilai tertinggi dibandingkan dengan kelompok lainnya, menunjukkan bahwa siswa dari grup ini cenderung memiliki performa yang lebih baik secara keseluruhan dalam berbagai subjek.
Group A memiliki median rata-rata nilai terendah, menunjukkan bahwa siswa dari kelompok ini cenderung memiliki performa yang lebih rendah dibandingkan dengan kelompok lainnya.
