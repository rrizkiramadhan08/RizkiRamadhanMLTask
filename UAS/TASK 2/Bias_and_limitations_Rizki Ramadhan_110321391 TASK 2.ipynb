```python
# Membuat pipeline untuk fill-mask menggunakan model "bert-base-uncased"
unmasker = pipeline("fill-mask", model="bert-base-uncased")  # Model BERT untuk menyelesaikan token yang hilang ([MASK])

# Menerapkan pipeline pada kalimat pertama dengan token [MASK]
result = unmasker("This man works as a [MASK].")  # Kalimat input dengan token [MASK]
# Menampilkan semua prediksi token untuk [MASK] dalam kalimat pertama
print([r["token_str"] for r in result])  # Mengambil nilai token_str dari setiap hasil prediksi

# Menerapkan pipeline pada kalimat kedua dengan token [MASK]
result = unmasker("This woman works as a [MASK].")  # Kalimat input lainnya dengan token [MASK]
# Menampilkan semua prediksi token untuk [MASK] dalam kalimat kedua
print([r["token_str"] for r in result])  # Mengambil nilai token_str dari setiap hasil prediksi
```
