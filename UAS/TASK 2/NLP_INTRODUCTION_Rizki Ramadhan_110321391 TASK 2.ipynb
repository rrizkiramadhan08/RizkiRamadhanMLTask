```python
# Instalasi library yang diperlukan: datasets, evaluate, dan transformers dengan opsi sentencepiece
!pip install datasets evaluate transformers[sentencepiece]
# Import library pipeline dari transformers
from transformers import pipeline
```

```python
# Pipeline untuk analisis sentimen
classifier = pipeline("sentiment-analysis")  # Membuat pipeline untuk analisis sentimen
print(classifier("I've been waiting for a HuggingFace course my whole life."))  # Menganalisis sentimen dari teks input
```

```python
classifier(
    ["I love you so much.", "I hate this so much!"] # Menganalisis sentimen dari teks input
)
```

```python
# Pipeline untuk klasifikasi tanpa pengawasan (zero-shot classification)
classifier = pipeline("zero-shot-classification")  # Membuat pipeline zero-shot classification
print(classifier(
    "This is a course about the Transformers library",  # Teks yang akan diklasifikasikan
    candidate_labels=["education", "politics", "business"],  # Label kandidat untuk klasifikasi
))
```

```python
# Pipeline untuk menghasilkan teks (text generation)
generator = pipeline("text-generation")  # Membuat pipeline untuk text generation
print(generator("In this course, we will teach you how to"))  # Menghasilkan teks berdasarkan input
```

```python
# Pipeline text generation dengan model spesifik (distilgpt2)
generator = pipeline("text-generation", model="distilgpt2")  # Menentukan model text generation
print(generator(
    "In this course, we will teach you how to",  # Input awal teks
    max_length=30,  # Panjang maksimum output teks
    num_return_sequences=2,  # Jumlah keluaran yang dihasilkan
))
```

```python
# Pipeline untuk melengkapi teks yang hilang (mask-filling)
unmasker = pipeline("fill-mask")  # Membuat pipeline untuk fill-mask
print(unmasker(
    "This course will teach you all about <mask> models.",  # Teks dengan token <mask> untuk dilengkapi
    top_k=2,  # Menampilkan 2 hasil teratas
))
```

```python
# Pipeline untuk Named Entity Recognition (NER)
ner = pipeline("ner", grouped_entities=True)  # Membuat pipeline untuk ekstraksi entitas nama
print(ner("My name is John and I work at Google in Bandung."))  # Ekstraksi entitas dari teks input
```

```python
# Pipeline untuk menjawab pertanyaan
question_answerer = pipeline("question-answering")  # Membuat pipeline untuk menjawab pertanyaan
print(question_answerer(
    question="what is my name?",  # Pertanyaan
    context="My name is John and I work at Google in Bandung",  # Konteks dari pertanyaan
))
```

```python
# Pipeline untuk merangkum teks
summarizer = pipeline("summarization")  # Membuat pipeline untuk summarization
print(summarizer(
    """
    America has changed dramatically during recent years. Not only has the number of
    graduates in traditional engineering disciplines such as mechanical, civil,
    electrical, chemical, and aeronautical engineering declined, but in most of
    the premier American universities engineering curricula now concentrate on
    and encourage largely the study of engineering science. As a result, there
    are declining offerings in engineering subjects dealing with infrastructure,
    the environment, and related issues, and greater concentration on high
    technology subjects, largely supporting increasingly complex scientific
    developments. While the latter is important, it should not be at the expense
    of more traditional engineering.

    Rapidly developing economies such as China and India, as well as other
    industrial countries in Europe and Asia, continue to encourage and advance
    the teaching of engineering. Both China and India, respectively, graduate
    six and eight times as many traditional engineers as does the United States.
    Other industrial countries at minimum maintain their output, while America
    suffers an increasingly serious decline in the number of engineering graduates
    and a lack of well-educated engineers.
    """
))
```

```python
# Pipeline untuk terjemahan teks
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-fr-en")  # Pipeline terjemahan dengan model spesifik
print(translator("Ce cours est produit par Hugging Face."))  # Menerjemahkan teks dari bahasa Prancis ke Inggris
```
