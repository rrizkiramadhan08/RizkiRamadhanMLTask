```python
# Menggunakan pipeline untuk fill-mask
from transformers import pipeline

# Membuat pipeline dengan model "camembert-base" untuk tugas fill-mask
camembert_fill_mask = pipeline("fill-mask", model="camembert-base")

# Menerapkan pipeline pada teks dengan token <mask>
results = camembert_fill_mask("Le camembert est <mask> :)")
```

```python
# Menampilkan hasil prediksi untuk token <mask>
print(results)
```

```python
# Menggunakan CamembertTokenizer dan CamembertForMaskedLM secara langsung
from transformers import CamembertTokenizer, CamembertForMaskedLM

# Memuat tokenizer dan model dari "camembert-base"
tokenizer = CamembertTokenizer.from_pretrained("camembert-base")
model = CamembertForMaskedLM.from_pretrained("camembert-base")
```

```python
# Alternatif dengan AutoTokenizer dan AutoModelForMaskedLM
from transformers import AutoTokenizer, AutoModelForMaskedLM

# AutoTokenizer secara otomatis mengenali tokenizer berdasarkan nama model
tokenizer = AutoTokenizer.from_pretrained("camembert-base")
# AutoModelForMaskedLM memuat model pralatih untuk Masked Language Modeling
model = AutoModelForMaskedLM.from_pretrained("camembert-base")
```
