```python
import gradio as gr

# Fungsi untuk memberikan salam
def greet(name):
    return "Hello " + name

# Membuat antarmuka Gradio untuk menyapa pengguna
demo = gr.Interface(
    fn=greet,  # Fungsi untuk memproses input
    inputs="text",  # Input berupa teks
    outputs="text"  # Output berupa teks
)

# Meluncurkan antarmuka
demo.launch()
```

```python
import gradio as gr

# Fungsi untuk memberikan salam
def greet(name):
    return "Hello " + name

# Membuat komponen Textbox dengan label dan placeholder
textbox = gr.Textbox(label="Type your name here:", placeholder="John Doe", lines=2)

# Membuat antarmuka Gradio dengan komponen Textbox sebagai input
gr.Interface(
    fn=greet,  # Fungsi untuk memproses input
    inputs=textbox,  # Input berupa Textbox
    outputs="text"  # Output berupa teks
).launch()
```

```python
from transformers import pipeline

# Memuat model pipeline untuk text-generation
model = pipeline("text-generation")

# Fungsi untuk menghasilkan teks berdasarkan prompt
def predict(prompt):
    completion = model(prompt)[0]["generated_text"]  # Menghasilkan teks berdasarkan input
    return completion
```

```python
# Membuat antarmuka Gradio untuk text-generation
gr.Interface(
    fn=predict,  # Fungsi untuk memproses input
    inputs="text",  # Input berupa teks
    outputs="text"  # Output berupa teks
).launch()
```
