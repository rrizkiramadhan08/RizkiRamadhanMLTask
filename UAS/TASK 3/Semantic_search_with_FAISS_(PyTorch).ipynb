```python
from datasets import load_dataset
# Memuat dataset GitHub Issues dari HuggingFace
issues_dataset = load_dataset("lewtun/github-issues", split="train")
issues_dataset
```

```python
# Memfilter dataset untuk hanya menyertakan issue (bukan pull request) dengan komentar
issues_dataset = issues_dataset.filter(
    lambda x: (x["is_pull_request"] == False and len(x["comments"]) > 0)
)
issues_dataset
```

```python
# Mendapatkan kolom dataset dan menentukan kolom yang akan disimpan
columns = issues_dataset.column_names
columns_to_keep = ["title", "body", "html_url", "comments"]
columns_to_remove = set(columns_to_keep).symmetric_difference(columns)
# Menghapus kolom yang tidak relevan dari dataset
issues_dataset = issues_dataset.remove_columns(columns_to_remove)
issues_dataset
```

```python
# Mengatur format dataset menjadi DataFrame pandas
issues_dataset.set_format("pandas")
df = issues_dataset[:]
```

```python
# Mengakses daftar komentar pada entri pertama
df["comments"][0].tolist()
```

```python

# Mengeksplorasi kolom "comments" sehingga setiap komentar menjadi baris baru
comments_df = df.explode("comments", ignore_index=True)
comments_df.head(4)
```

```python
# Mengonversi DataFrame menjadi dataset HuggingFace
from datasets import Dataset

comments_dataset = Dataset.from_pandas(comments_df)
comments_dataset
```

```python
# Menambahkan panjang komentar sebagai kolom baru
comments_dataset = comments_dataset.map(
    lambda x: {"comment_length": len(x["comments"].split())}
)
```

```python
# Memfilter dataset untuk hanya menyertakan komentar dengan panjang lebih dari 15 kata
comments_dataset = comments_dataset.filter(lambda x: x["comment_length"] > 15)
comments_dataset
```

```python
# Mendefinisikan fungsi untuk menggabungkan teks dari `title`, `body`, dan `comments`
def concatenate_text(examples):
    return {
        "text": examples["title"]
        + " \n "
        + examples["body"]
        + " \n "
        + examples["comments"]
    }

# Menambahkan kolom "text" yang merupakan gabungan dari `title`, `body`, dan `comments`
comments_dataset = comments_dataset.map(concatenate_text)
```

```python
# Memuat model dan tokenizer dari `sentence-transformers`
from transformers import AutoTokenizer, AutoModel

model_ckpt = "sentence-transformers/multi-qa-mpnet-base-dot-v1"
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
model = AutoModel.from_pretrained(model_ckpt)
```

```python
# Mengecek apakah GPU tersedia, jika tidak gunakan CPU
import torch

# Mengecek apakah GPU tersedia, jika tidak, gunakan CPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Memindahkan model ke perangkat yang sesuai
model.to(device)

print(f"Model menggunakan perangkat: {device}")
```

```python
# Mendefinisikan fungsi pooling untuk mengambil representasi CLS token
def cls_pooling(model_output):
    return model_output.last_hidden_state[:, 0]
```

```python
# Mendefinisikan fungsi untuk mendapatkan embedding dari teks
def get_embeddings(text_list):
    encoded_input = tokenizer(
        text_list, padding=True, truncation=True, return_tensors="pt"
    )
    encoded_input = {k: v.to(device) for k, v in encoded_input.items()}
    model_output = model(**encoded_input)
    return cls_pooling(model_output)
```

```python
# Menghasilkan embedding untuk entri pertama dari komentar
embedding = get_embeddings(comments_dataset["text"][0])
embedding.shape
```

```python

# Menambahkan kolom embedding ke dataset
embeddings_dataset = comments_dataset.map(
    lambda x: {"embeddings": get_embeddings(x["text"]).detach().cpu().numpy()[0]}
)
```

```python
# Menambahkan indeks Faiss untuk kolom embedding
embeddings_dataset.add_faiss_index(column="embeddings")
```

```python
# Mendefinisikan pertanyaan yang akan dicocokkan dengan komentar di dataset
question = "How can I load a dataset offline?"
question_embedding = get_embeddings([question]).cpu().detach().numpy()
question_embedding.shape
```

```python
# Mendapatkan komentar yang paling relevan dengan pertanyaan menggunakan Faiss
scores, samples = embeddings_dataset.get_nearest_examples(
    "embeddings", question_embedding, k=5
)
```

```python
# Mengonversi hasil pencocokan ke DataFrame untuk analisis lebih lanjut
import pandas as pd

samples_df = pd.DataFrame.from_dict(samples)
samples_df["scores"] = scores
samples_df.sort_values("scores", ascending=False, inplace=True)
```

```python

# Mencetak hasil pencocokan dengan informasi komentar, skor, judul, dan URL
for _, row in samples_df.iterrows():
    print(f"COMMENT: {row.comments}")
    print(f"SCORE: {row.scores}")
    print(f"TITLE: {row.title}")
    print(f"URL: {row.html_url}")
    print("=" * 50)
    print()
```
