```python
# Mengimpor tokenizer dari pustaka transformers
from transformers import AutoTokenizer
# Memuat tokenizer berbasis model "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
# Menampilkan tipe backend tokenizer yang digunakan (misalnya, tokenizer cepat atau biasa)
print(type(tokenizer.backend_tokenizer))
```

```python
# Menggunakan normalizer dari backend tokenizer untuk menormalisasi string
# Normalisasi ini melibatkan langkah seperti menghilangkan diakritik atau mengubah teks ke huruf kecil
print(tokenizer.backend_tokenizer.normalizer.normalize_str("Héllò hôw are ü?"))
```

```python
# Menggunakan pre-tokenizer untuk memisahkan string menjadi token awal
# Pre-tokenizer akan memisahkan teks berdasarkan spasi, tanda baca, dll.
tokenizer.backend_tokenizer.pre_tokenizer.pre_tokenize_str("Hello, how are  you?")
```

```python
# Memuat tokenizer berbasis model "gpt2"
tokenizer = AutoTokenizer.from_pretrained("gpt2")
# Menggunakan pre-tokenizer dari tokenizer "gpt2" untuk memisahkan string
# Pre-tokenizer model ini akan menangani teks berbeda sesuai dengan arsitektur model GPT-2
tokenizer.backend_tokenizer.pre_tokenizer.pre_tokenize_str("Hello, how are  you?")
```

```python
# Memuat tokenizer berbasis model "t5-small"
tokenizer = AutoTokenizer.from_pretrained("t5-small")
# Menggunakan pre-tokenizer dari tokenizer "t5-small" untuk memisahkan string
# Pre-tokenizer model ini bekerja berbeda dibandingkan model sebelumnya, tergantung pada spesifikasi T5
tokenizer.backend_tokenizer.pre_tokenizer.pre_tokenize_str("Hello, how are  you?")
```
