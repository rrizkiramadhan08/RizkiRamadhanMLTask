# Multiple Linear Regression

## Importing the libraries

```
# Importing the libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
```

## Importing the dataset

```
dataset = pd.read_csv('/content/50_Startups.csv')
```

```
X = dataset.iloc[:, :-1].values    # Memisahkan semua kolom kecuali kolom terakhir (variabel independen).
y = dataset.iloc[:, -1].values    # Memisahkan kolom terakhir (Profit) sebagai variabel dependen.
```

## Feature Engineering

```
# Encoding categorical data (State column)
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

## Splitting the dataset into the Training set and Test set

```
# Splitting the dataset into the Training set and Test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Training the Multiple Linear Regression model on the Training set

```
# Training the Multiple Linear Regression model on the Training set
regressor = LinearRegression()
regressor.fit(X_train, y_train) # Melatih model menggunakan set pelatihan (X_train, y_train).
```

## Predicting the Test set results

```
# Predicting the Test set results
y_pred = regressor.predict(X_test) # Menggunakan model yang sudah dilatih untuk memprediksi hasil dari set pengujian (X_test).

# Creating a DataFrame to compare the actual and predicted values
comparison = pd.DataFrame({'Actual Profit': y_test, 'Predicted Profit': y_pred})

# Display the first few rows of the comparison
print(comparison.head())

# Optionally: Calculate model performance metrics
from sklearn.metrics import mean_absolute_error, r2_score

mae = mean_absolute_error(y_test, y_pred)  # Mean Absolute Error
r2 = r2_score(y_test, y_pred)  # R-squared (coefficient of determination)

print(f'Mean Absolute Error: {mae}')
print(f'R-squared: {r2}')
```
